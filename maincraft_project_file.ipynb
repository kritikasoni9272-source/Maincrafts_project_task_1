```python
# Student Performance Analysis

This project analyzes student academic performance using Python.
The objective is to examine:-
- The average final grade (G3).
- Number of high-performing students.
- Relationship between study time and grades.
- Gender-wise performance comparison.
```

```python
# importing the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Data Loading

The dataset is loaded using Pandas. The separator used is ';' as the file is semicolon-separated.
```

```python
# Reading Data

df=pd.read_csv('student-mat.csv', sep=';')
print("First Five rows of the Dataset : ")
print(df.head())
```

```python
## Data Exploration and Cleaning

In this step:
- Missing values are checked
- Duplicate rows are identified
- Dataset shape is verified
- Data types are inspected
```

```python
# exlpore and clean data sets

print("\nMisisng values in each columns : ")
print(df.isnull().sum())
print("\nNumber of duplicated rows : ")
print(df.duplicated().sum())
print("Shape of the Dataset : ")
print(df.shape)
print("Data types of each columns : ")
print(df.dtypes)
```

```python
## Statistical Analysis

The following metrics are calculated:
- Average final grade (G3)
- Number of students scoring above 15
- Correlation between study time and performance
- Gender-wise average performance
```

```python
# Analysis

report={
    "Average final grade (G3) ":df['G3'].mean(),
    "Students scored above 15 ":(df['G3'] > 15).sum(),
    "Correlation of study time with performance ":df['studytime'].corr(df['G3'])
}
print("\nOverall Analysis Report : ")
for key, value in report.items():
    print(f"{key} : {value}")

print('\nAverage performance result by gender:')
gender_perf=df.groupby("sex")['G3'].mean()
print(gender_perf)
```

```python
## Distribution of Final Grades

This histogram shows how student final grades are distributed across the dataset.
```

```python
# Visualization

plt.figure()
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.tight_layout()
plt.savefig('Distribution of Final Grades', dpi=300, bbox_inches='tight')
plt.show()
```

```python
### Insight

Most students are clustered in the mid-grade range, 
with fewer students achieving extremely high or low scores.
```

```python
## Study Time vs Final Grade

This scatter plot examines the relationship between study time and final performance.
```

```python
sns.scatterplot(x='studytime', y='G3', data=df, legend=True, color='teal', s=80, alpha=0.7)
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grades")
plt.tight_layout()
plt.savefig('Study Time vs Final Grades', dpi=300, bbox_inches='tight')
plt.show()
```

```python
### Insight

There is a positive relationship between study time and final grade, 
indicating that increased study effort is generally associated with better performance.
```

```python
## Average Final Grade by Gender

This bar chart compares the average final grades of male and female students.
```

```python
sns.barplot(x='sex', y='G3', data=df, errorbar=None, hue='sex', palette={'F':'lightpink','M':'skyblue'}, legend=True)
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Average Final Grade by Gender")
plt.tight_layout()
plt.savefig('Average Final Grade by Gender', dpi=300, bbox_inches='tight')
plt.show()
```

```python
### Insight

The visualization shows a slight difference in average performance between genders.
```

```python
## Final Conclusion

- The average final grade provides a baseline academic performance measure.
- A subset of students scored above 15, indicating strong academic achievers.
- Study time shows a positive correlation with performance.
- Gender-based differences exist but are not significantly large.

Overall, academic performance appears to be influenced by study habits and demographic factors.
```
